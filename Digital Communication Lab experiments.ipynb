### Generate Sine and Cosine Signals
numpy and matplotlib libraries are required

```python
import numpy as np
import matplotlib.pyplot as plt
t= np.arange(0,1,0.01)
f=2
x = np.sin(2*np.pi*f*t)
plt.stem(t,x,'r')
plt.show()
y= np.cos(2*np.pi*f*t)
plt.stem(t,y,'b')
plt.show()
```

Generate unipolar and polar signals

numpy and matplotlib libraries are required


```python
import numpy as np
import matplotlib.pyplot as plt
unipolardata=np.array([1,0, 1, 1, 0])
x= unipolardata.size
samples= 500
t= np.linspace(0,5,samples)
unipolarwave= np.repeat(unipolardata,samples/x )
plt.plot(unipolarwave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('Unipolar NRZ')
plt.grid()
plt.show()

polardata= 2*unipolardata-1
polarwave= np.repeat(polardata,samples/x )
plt.plot(polarwave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('Polar NRZ')
plt.grid()
plt.show()
```
